# Data preparation

```python
import pandas as pd
import os
```

## Monthly Detection Data  [3 provinces]

```python
df_detection = pd.read_csv(os.path.join('data','Breeding Site Detection','detection.csv'))
df_detection['addrcode'] = df_detection['addrcode'].astype(str)
df_detection.head()
```

```python
len(df_detection)
```

### Bangkok Detection Data 

```python
df_detection_bangkok = df_detection.loc[df_detection['addrcode'].str[:2] == '10']
df_detection_bangkok.head()
```

```python
len(df_detection_bangkok)
```

### Nakhon-Si-Thammarat Detection Data

```python
df_detection_nakhon = df_detection.loc[df_detection['addrcode'].str[:2] == '80']
df_detection_nakhon.head()
```

```python
len(df_detection_nakhon)
```

### Krabi Detection Data

```python
df_detection_krabi = df_detection.loc[df_detection['addrcode'].str[:2] == '81']
df_detection_krabi.head()
```

```python
len(df_detection_krabi)
```

## Google Street View Coverage Area

```python
df_gsv_coverage = pd.read_csv(os.path.join('data','Area and GSV coverage','gsv_coverage.csv'))
df_gsv_coverage['addrcode'] = df_gsv_coverage['addrcode'].astype(str)
df_gsv_coverage.head()
```

## Area Data

```python
df_area = pd.read_csv(os.path.join('data','Area and GSV coverage','area.csv'))
df_area = df_area.rename(columns = {'subdist_ID' : 'addrcode'})
df_area['addrcode'] = df_area['addrcode'].astype(str)
df_area.head()
```

## Monthly Dengue cases Data
### Normalized by Population x 10000

```python
df_dengue_cases = pd.read_csv(os.path.join('data','Dengue Case','dengue_cases_final.csv'))
df_dengue_cases = df_dengue_cases.drop("Unnamed: 0", axis=1)
df_dengue_cases.head()
```
