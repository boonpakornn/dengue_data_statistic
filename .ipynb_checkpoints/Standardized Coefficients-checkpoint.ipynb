# Standardized Coefficient Data

```python
import pandas as pd
import os
import numpy as np
```

```python
province = 'bangkok'
```

```python
df_df = pd.read_csv(os.path.join('Data','Standardized Coefficient/DF_monthly.csv'), index_col = 0)
#df_df = df_df.drop(['NumberCases'],axis = 1)
df_df['addrcode'] = df_df['addrcode'].astype(str)
df_df['Year'] = df_df['Year'].astype(str)
df_df['Month'] = df_df['Month'].astype(str)
if province == 'bangkok':
    df_df = df_df.loc[df_df['addrcode'].str[:2] == '10']
elif province == 'nst':
    df_df = df_df.loc[df_df['addrcode'].str[:2] == '80']
elif province == 'krabi':
    df_df = df_df.loc[df_df['addrcode'].str[:2] == '81']
list = ['2014','2018']
for i in range(len(list)):
    df_df = df_df[df_df.Year != list[i]]
df_df = df_df.groupby([df_df['addrcode'],df_df['Year']]).sum().reset_index()
df_df.head()
```

```python
df_pop = pd.read_csv(os.path.join('Data','Population/addrcode-index/Population.csv'))
df_pop['addrcode'] = df_pop['addrcode'].astype(str)
df_pop.head()
```

```python
df_df = pd.merge(df_df, df_pop, how = 'inner')
df_df['Cases'] = df_df['Cases']/ (df_df['Total']/10000)
df_df = df_df.drop('Total', axis =1)
df_df.head()
```

```python
## inserting RF and LST
df_lst = pd.read_csv(os.path.join('Data','Climatic Data','LST_subdist_monthly.csv'))
df_lst = df_lst.drop('Unnamed: 0', axis =1 )
df_lst['addrcode'] = df_lst['addrcode'].astype(str)
df_lst['Year'] = df_lst['Year'].astype(str)
df_lst['Month'] = df_lst['Month'].astype(str)
df_lst.head()
```

```python
df_lst_min = df_lst.groupby([df_lst['addrcode'], df_lst['Year']]).min().reset_index()
df_lst_min = df_lst_min.drop('Month', axis = 1)
df_lst_min.columns = [['addrcode','Year','MIN_LST']]
df_lst_min.head()
```

```python
df_lst_max = df_lst.groupby([df_lst['addrcode'],df_lst['Year']]).max().reset_index()
df_lst_max = df_lst_max.drop('Month', axis = 1)
df_lst_max.columns = [['addrcode','Year','MAX_LST']]
df_lst_max.head()
```

```python
df_lst_yearly = pd.read_csv(os.path.join('Data','Climatic Data','LST_subdist_yearly.csv'), index_col = 0)
df_lst_yearly['addrcode'] = df_lst_yearly['addrcode'].astype(str)
df_lst_yearly['Year'] = df_lst_yearly['Year'].astype(str)
#df_lst_yearly['Month'] = df_lst_yearly['Month'].astype(str)
df_lst_yearly.columns = [['addrcode','Year','AVG_LST']]
df_lst_yearly.head()
```

```python
df_std_co['Bin'] = ((df_std_co['Bin'])/ np.std(df_std_co['Bin'])).abs()
df_std_co['Bowl'] = ((df_std_co['Bowl'])/ np.std(df_std_co['Bowl'])).abs()
df_std_co['Bucket'] = ((df_std_co['Bucket'])/ np.std(df_std_co['Bucket'])).abs()
df_std_co['Misc_Short'] = ((df_std_co['Misc_Short'])/ np.std(df_std_co['Misc_Short'])).abs()
df_std_co['Jar'] = ((df_std_co['Jar'])/ np.std(df_std_co['Jar'])).abs()
df_std_co['Pottedplant'] = ((df_std_co['Pottedplant'])/ np.std(df_std_co['Pottedplant'])).abs()
df_std_co['Tire'] = ((df_std_co['Tire'])/ np.std(df_std_co['Tire'])).abs()
df_std_co['Misc_Tall'] = ((df_std_co['Misc_Tall'])/ np.std(df_std_co['Misc_Tall'])).abs()
df_std_co['Total'] = ((df_std_co['Total'])/ np.std(df_std_co['Total'])).abs()
df_std_co.head()
```
