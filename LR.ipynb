```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
import os
import matplotlib.pyplot as plt
```

```python
# Load the Dengue dataset
#Dengue = datasets.load_Dengue()
df_train =  pd.read_csv(os.path.join('Data','Linear Regression','8022_sum_10000pop.csv'), header=0, skiprows=0)
#X = Dengue.iloc[:,range(4,18)]
X = Dengue.iloc[:,range(4,18)]
Y = Dengue.iloc[:,[3]]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
print('Linear Regression R squared: %.4f' % regressor.score(X_test, y_test))

# calculate RMSE
lin_mse = mean_squared_error(y_pred, y_test)
lin_rmse = np.sqrt(lin_mse)
print('Linear Regression RMSE: %.4f' % lin_rmse)

# calculate MAE
from sklearn.metrics import mean_absolute_error
lin_mae = mean_absolute_error(y_pred, y_test)
print('Linear Regression MAE: %.4f' % lin_mae)
print('\n')

y_test = y_test.reset_index()
y_test = y_test.drop('index', axis = 1)
predicted = pd.DataFrame({'Predicted':y_pred[:,0]})
df_compare = pd.concat([predicted,y_test], axis = 1)
df_compare.columns = ['predicted', 'actual']
plot = plt.plot(df_compare)
plot = plt.ylabel('Dengue incidences per 10,000 population')
plot = plt.xlabel('Week number')
plot = plt.legend(['predicted', 'actual'])
plot = plt.title('Chang Klang District')
```
