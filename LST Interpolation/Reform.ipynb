```python
import pandas as pd
import os
```

```python
df = pd.read_csv('LSTweekly_NST_2017.csv')
df = df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
df = df.set_index(['Weekno'])
df.columns = df.columns.str.extract('(\d+)', expand=False)
df = df.stack().reset_index(name='temperature').rename(columns={'level_1':'addrcode'})
df['Weekno'] = df['Weekno'].astype(str)
df.head()
```

```python
df['Week'] = df['Weekno'].str[:-4]
df['Year'] = df['Weekno'].str[-4:]
df = df.drop('Weekno', axis = 1)
df = df[['Week','Year','addrcode','temperature']]
df = df.sort_values(['Week','Year','addrcode'])
df.head(5)
```

```python
df.to_csv('LST_weekly_NST_2017.csv', encoding = 'utf-8')
```
