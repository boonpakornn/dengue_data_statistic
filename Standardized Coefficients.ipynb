```python
import pandas as pd
import os
import numpy as np
```

```python
province = 'bangkok'
```

```python
df_df = pd.read_csv(os.path.join('Data','Standardized Coefficient/DF_standardized.csv'), index_col = 0)
df_df = df_df.drop(['NumberCases'],axis = 1)
df_df['addrcode'] = df_df['addrcode'].astype(str)
df_df.head()
```

```python
df_df = df_df.loc[df_df['addrcode'].str[:2] == '10']
df_df.head()
```

```python
df_cd = pd.read_csv(os.path.join('Data','Container Density','Subdistrict Level','AdjustedCD_'+province+'.csv'),index_col = 0)
df_cd['addrcode'] = df_cd['addrcode'].astype(str)
df_cd.head()
```

```python
df_std_co = pd.merge(df_df, df_cd, how = 'inner')
df_std_co.head()
```

```python
df_std_co['Bin'] = ((df_std_co['Bin'] - np.mean(df_std_co['Bin']))/ np.std(df_std_co['Bin'])).abs()
df_std_co['Bowl'] = ((df_std_co['Bowl'] - np.mean(df_std_co['Bowl']))/ np.std(df_std_co['Bowl'])).abs()
df_std_co['Bucket'] = ((df_std_co['Bucket'] - np.mean(df_std_co['Bucket']))/ np.std(df_std_co['Bucket'])).abs()
df_std_co['Misc Short'] = ((df_std_co['Misc Short'] - np.mean(df_std_co['Misc Short']))/ np.std(df_std_co['Misc Short'])).abs()
df_std_co['Jar'] = ((df_std_co['Jar'] - np.mean(df_std_co['Jar']))/ np.std(df_std_co['Jar'])).abs()
df_std_co['Pottedplant'] = ((df_std_co['Pottedplant'] - np.mean(df_std_co['Pottedplant']))/ np.std(df_std_co['Pottedplant'])).abs()
df_std_co['Tire'] = ((df_std_co['Tire'] - np.mean(df_std_co['Tire']))/ np.std(df_std_co['Tire'])).abs()
df_std_co['Misc Tall'] = ((df_std_co['Misc Tall'] - np.mean(df_std_co['Misc Tall']))/ np.std(df_std_co['Misc Tall'])).abs()
df_std_co['Total'] = ((df_std_co['Total'] - np.mean(df_std_co['Total']))/ np.std(df_std_co['Total'])).abs()
df_std_co.head()
```

```python
df_std_co.to_csv('Data/Standardized Coefficient/Std_'+ province+'.csv', encoding = 'utf-8')
```
