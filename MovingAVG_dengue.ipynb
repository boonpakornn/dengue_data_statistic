## 2 variables

```python
df_dengue = pd.read_csv(os.path.join('Data','Modeling','Training&Testing','Bangkok','Original','train_bangkok_dist_total.csv'))
df_dengue = df_dengue.drop('Unnamed: 0', axis = 1)
df_dengue['addrcode'] = df_dengue['addrcode'].astype(str)
df_dengue['Year'] = df_dengue['Year'].astype(str)
df_dengue.head()
```

```python
movavg = []
movavg.append(df_dengue['DFp10000Pop'][0])
```

```python
for i in range(len(df_dengue)-1):
    tmp = (df_dengue['DFp10000Pop'][i]+df_dengue['DFp10000Pop'][i+1])/2
    movavg.append(tmp)
```

```python
df_dengue['DF_MA'] = movavg
```

```python
df_dengue.head()
```

```python
df_dengue = df_dengue[['addrcode','Week','Year','DFp10000Pop','DF_MA','rainfall','temperature','bin_pop9s','bowl_pop9s','bucket_pop9s','misc_short_pop9s','jar_pop9s','pottedplant_pop9s','tire_pop9s','misc_tall_pop9s','total_pop9s']]
df_dengue.columns = ['addrcode','Week','Year','DF_0','DFma_0','rainfall','temperature','bin_pop9s','bowl_pop9s','bucket_pop9s','misc_short_pop9s','jar_pop9s','pottedplant_pop9s','tire_pop9s','misc_tall_pop9s','total_pop9s']
df_dengue.head()
```

```python
df_dengue.to_csv(os.path.join('Data','Modeling','Training&Testing','Bangkok','train_bangkok_dist_total_mavg2.csv'), encoding = 'utf-8')
```
