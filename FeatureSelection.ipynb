```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
import os
import matplotlib.pyplot as plt
```

```python
# Load the Dengue dataset
#Dengue = datasets.load_Dengue()
df_train =  pd.read_csv(os.path.join('Data','Feature Selection','train_lag.csv'), header=0, skiprows=0)
df_train = df_train.drop('Unnamed: 0', axis = 1)
df_test =  pd.read_csv(os.path.join('Data','Feature Selection','test_lag.csv'), header=0, skiprows=0)
df_test = df_test.drop('Unnamed: 0', axis = 1)
index = range(15)
df_result = pd.DataFrame(index = index,columns=['Deleted', 'Rsquared', 'RMSE', 'MAE', 'AIC'])
```

```python
columns = df_train.columns.values.tolist()
#name = [100]
#for i in range(15):
#    name.append(columns[4+i])
    
for i in range(15):
    df_result['Deleted'][i] = columns[4+i]
    x_train = df_train.iloc[:,range(4,34)]
    x_train = x_train.drop(x_train.columns[i], axis=1)
    y_train = df_train.iloc[:,[3]]
    x_test = df_test.iloc[:,range(4,34)]
    x_test = x_test.drop(x_test.columns[i], axis=1)
    y_test = df_test.iloc[:,[3]]

    regressor = LinearRegression()
    regressor.fit(x_train, y_train)
    y_pred = regressor.predict(x_test)
    #print('Linear Regression R squared: %.4f' % regressor.score(x_test, y_test))
    df_result['Rsquared'][i] = regressor.score(x_test, y_test)

    # calculate RMSE
    lin_mse = mean_squared_error(y_pred, y_test)
    lin_rmse = np.sqrt(lin_mse)
    #print('Linear Regression RMSE: %.4f' % lin_rmse)
    df_result['RMSE'][i] = lin_mse

    # calculate MAE
    from sklearn.metrics import mean_absolute_error
    lin_mae = mean_absolute_error(y_pred, y_test)
    #print('Linear Regression MAE: %.4f' % lin_mae)
    df_result['MAE'][i] = lin_mae

    # calculate AIC
    resid = y_test - y_pred
    sse = sum(resid.values**2)
    k = 31
    AIC= 2*k - 2*np.log(sse)
    #print('Linear Regression AIC : %.4f' % AIC)
    #print('\n')
    df_result['AIC'][i] = AIC[0]
```
